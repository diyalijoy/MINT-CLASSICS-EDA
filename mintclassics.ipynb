### MINT CLASSICS DATABASE EDA

Here, we have a fake database of a real company 'Mint Classics'. We will use MySQL Workbench to do exploratory data analysis on the database. Our objective is to come up with useful insights about the operations of this company and visualize the data using Tableau.

For my reference and use in analysis, the EER (Extended Entity-Relationship) diagram that models the structure of the Mint Classics database was provided here.
